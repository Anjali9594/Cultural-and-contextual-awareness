```python
import cv2
import numpy as np
from keras.models import model_from_json

# Load the trained emotion detection model
json_file = open("C:/Users/Anjali/Desktop/streaming video project 5/emotion_model.json", "r")
loaded_model_json = json_file.read()
json_file.close()
emotion_model = model_from_json(loaded_model_json)
emotion_model.load_weights("C:/Users/Anjali/Desktop/streaming video project 5/emotion_model.weights.h5")

# Emotion labels
emotion_dict = {0: 'Angry', 1: 'Disgust', 2: 'Fear', 3: 'Happy', 4: 'Sad', 5: 'Surprise', 6: 'Neutral'}

# Open the video capture
cap = cv2.VideoCapture("C:/Users/Anjali/Desktop/streaming video project 5/sample video/sample video2.webm")

while True:
    ret, frame = cap.read()
    
    if not ret:
        break
    
    # Detect faces in the frame
    # Use absolute file path for Haar cascade classifier
    face_cascade = cv2.CascadeClassifier('C:/Users/Anjali/Desktop/streaming video project 5/haarcascade_frontalface_default.xml')

    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray_frame, scaleFactor=1.3, minNeighbors=5)
    
    for (x, y, w, h) in faces:
        # Extract the face ROI and preprocess for prediction
        face_roi = gray_frame[y:y + h, x:x + w]
        cropped_img = np.expand_dims(np.expand_dims(cv2.resize(face_roi, (48, 48)), -1), 0)
        
        # Predict emotion
        emotion_prediction = emotion_model.predict(cropped_img)
        max_index = np.argmax(emotion_prediction)
        emotion_label = emotion_dict[max_index]
        
        # Display the emotion label on the frame
        cv2.putText(frame, emotion_label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36, 255, 12), 2)
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)
    
    # Display the frame with emotion detection
    cv2.imshow('Emotion Detection', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
